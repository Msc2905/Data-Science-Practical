```python
book1 = book.drop(['Unnamed: 0'], axis = 1)
book1.tail(2)
```

```python
book = book1.rename({'User.ID':'UserId','Book.Title':'Title','Book.Rating':'Rating'}, axis=1)
book
```

```python
book_new = book.pivot_table(index='UserId', columns='Title', values='Rating')
book_new
```

```python
book_new.fillna(0, inplace=True)
book_new.index = book['UserId'].unique()
book_new
```

```python
user = 1 - pairwise_distances( book_new.values, metric='cosine')
user
```

```python
np.fill_diagonal(user, 0)
user1 = pd.DataFrame(user)
user1
```

```python
user1.index = book.UserId.unique()
user1.columns = book.UserId.unique()
user1
```

```python
def recommend(custID):
    simID = user1.loc[custID].idxmax()
    simID_book = book[book['UserId'] == simID].Title
    custID_book = book[book['UserId'] == custID].Title
    return set(simID_book) - set(custID_book)
```

```python
recommend(276813)
```
